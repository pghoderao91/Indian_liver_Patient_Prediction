```python
## Creating Independent and Targeted Variable

x= data.drop(['Target','AB','TP','Ap'],axis=1)
y= data['Target']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

**Balancing The Data**

**Smote:** SMOTE (synthetic minority oversampling technique) is one of the most commonly used oversampling methods to solve the imbalance problem. It aims to balance class distribution by randomly increasing minority class examples by replicating them. SMOTE synthesises new minority instances between existing minority instances.

```python
! pip install imblearn
```

```python
# Balance the data
from collections import Counter
from imblearn.over_sampling import SMOTE

sm = SMOTE()
print(Counter(y_train))
x_sm,y_sm = sm.fit_resample(x_train,y_train)
print(Counter(y_sm))
```

```python
print('x_train.shape:',x_train.shape)
print('x_test.shape:',x_test.shape)
print('y_train.shape:',y_train.shape)
print('y_test.shape:',y_test.shape)
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(x_sm,y_sm)

y_pred_test = lr.predict(x_test)

y_pred_train = lr.predict(x_sm)
```

**Evaluation of Training Accuracy**

```python
# importing library
from sklearn.metrics import accuracy_score,recall_score,f1_score,classification_report
log_train_accuracy = accuracy_score(y_pred_train,y_sm)

print("Training accuracy of Logistic regression model",log_train_accuracy*100)
print("Logistic regression training Classification report: \n",classification_report(y_pred_train,y_sm))
```

**Evaluation Testing Accuracy**

```python
log_test_accuracy = accuracy_score(y_pred_test,y_test)
print("Testing accuracy of Logistic regession model",log_test_accuracy*100)
print("Logistic regression testing Classification report: \n",classification_report(y_pred_test,y_test))
```

```python
# In heaert disease prediction recall is more important
recall_log = recall_score(y_pred_test,y_test)
recall_log*100
```

```python
pd.crosstab(y_pred_test,y_test)
```

**Hyper Parametertunning**

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start=100 ,stop=2000, num=10)] #No of decision tree in forest
max_features = ['auto', 'sqrt'] #Max no of feature consider to create decision tree
max_depth    = [int(x) for x in np.linspace(10,100,num=11)] #Max no of level in each decision tree
max_depth.append(None)
min_samples_split = [2,3,5,10] #Min number of data points placed in a node before the node is split
min_samples_leaf  = [1,2,3,4,5]  #Min number of data point allowed in leaf node

# Step:2 Creating dictionary of paramter
random_grid = {'n_estimators': n_estimators, 'max_features': max_features,
               'max_depth': max_depth, 'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

# Step:3 Creation of ofbject

rf_clf = RandomForestClassifier(random_state=42)

# Step:4 Create randomsearch with Parameter

rf_cv = RandomizedSearchCV(estimator=rf_clf, param_distributions=random_grid, n_iter=100, cv=3, verbose=2, random_state=42, n_jobs=-1,scoring='f1')

## Step:5 fitting Training Data

rf_cv.fit(x_sm,y_sm)

# Step6 : get best parameter

print("Best parameters:", rf_cv.best_params_)
```

```python
# Step:7 Create object and place the best paramter
rf_clf1 = RandomForestClassifier(** rf_cv.best_params_)

# Step:8 Fitting the training data
rf_clf1.fit(x_sm,y_sm)

# Step:9 Prediction on test data
rf_clf1_predict = rf_clf1.predict(x_test)
```

```python
rf_accuracy = accuracy_score(rf_clf1_predict,y_test)
print("Accuracy after hyperparameter tunning",rf_accuracy*100)
print("Classification report: \n",classification_report(rf_clf1_predict,y_test))
```

```python
rf_recall_score = recall_score(rf_clf1_predict,y_test)
print("Recall score:",rf_recall_score*100)
```

**Hyperparameter Tunning**

```python
from sklearn.model_selection import GridSearchCV

# Step 1
#creating dictionary--> key value pair of hyperparameters having key as parameter and values as its values
params = {
    "criterion":("gini", "entropy"), #quality of split
    "splitter":("best", "random"), # searches the features for a split
    "max_depth":(list(range(1, 20))), #depth of tree range from 1 to 19
    "min_samples_split":[2, 3, 4],    #the minimum number of samples required to split internal node
    "min_samples_leaf":list(range(1, 20)),#minimum number of samples required to be at a leaf node,we are passing list which is range from 1 to 19
}

## step 2 object creation

dt_clf = DecisionTreeClassifier(random_state=42)

## Step 3 create randomized search cv
dt_rcv = GridSearchCV(estimator=dt_clf,param_grid=params,scoring="f1",cv=4,verbose=2,n_jobs=-1)

## step 4 fitting the data in gridsearch cv

dt_rcv.fit(x_sm,y_sm)

# Step6 : get best parameter

print("Best parameters:", dt_rcv.best_params_)
```
